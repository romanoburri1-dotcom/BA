# This notebook is used to preprocess meteorological data from MeteoSchweiz from 01.01.2020 until 31.12.2025 for the station Zurich Kloten (KLO). The data has been downloaded previously from the MeteoSchweiz website in 10 minute intervals.

## Step 1: Import all necessary libraries

```python
import pandas as pd
```

Step 2: Import MeteoSchweiz csv

```python
meteo = pd.read_csv("data/raw/meteo_2020_2025.csv",
                    delimiter=';',
                    parse_dates=['reference_timestamp'],
                    dayfirst=True,
                    )
```

Set reference_timestamp to datetime format

```python
meteo["reference_timestamp"] = (
    pd.to_datetime(
        meteo["reference_timestamp"],
        dayfirst=True,
        utc=True
    )
    .astype("timestamp[ns, tz=UTC][pyarrow]")
)
```

Converting column "tre200s0", which is the current air temperature 2m above ground,  to numeric values and discard non existing values (NaN)

```python
meteo["tre200s0"] = pd.to_numeric(meteo["tre200s0"], errors="coerce")
```

Exporting the meteorological data as a pkl file

```python
output_path = "data/processed/meteo_processed.pkl"

# Save dataframe to pkl.
meteo.to_pickle(output_path)

print(f"File \"meteo_processed\" saved to: {output_path}")
```
