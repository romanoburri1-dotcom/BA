# This notebook is used to preprocess meteorological data from MeteoSchweiz from 01.01.2020 until 31.12.2025 for the station Zurich Kloten (KLO). The data has been downloaded previously from the MeteoSchweiz website in 10 minute intervals.

## Step 1: Import all necessary libraries

```python
import pandas as pd
```

Step 2: Import MeteoSchweiz csv

```python
meteo = pd.read_csv("data/raw/meteo_2020_2025.csv",
                    delimiter=';',
                    parse_dates=['reference_timestamp'],
                    dayfirst=True,
                    )
```

Set reference_timestamp datetime format

```python
meteo["reference_timestamp"] = (
    pd.to_datetime(
        meteo["reference_timestamp"],
        dayfirst=True,
        utc=True
    )
    .astype("datetime64[ns, UTC]")
)
```

Converting column "tre200s0", which is the current air temperature 2m above ground,  to numeric values and discard non existing values (NaN)

```python
meteo["tre200s0"] = pd.to_numeric(meteo["tre200s0"], errors="coerce")
```

Converting column "tde200s0", which is the current air dewpoint 2m above ground,  to numeric values and discard non existing values (NaN)

```python
meteo["tde200s0"] = pd.to_numeric(meteo["tde200s0"], errors="coerce")
```

Calculating spread and adding it as a new column into the dataframe

```python
meteo["spread"] = (meteo["tre200s0"] - meteo['tde200s0']).round(1)
```

Exporting the meteorological data as a parquet file

```python
output_path = "data/processed/meteo_processed.parquet"

# Save dataframe to parquet.
meteo.to_parquet(output_path, index=False)

print(f"File \"meteo_processed\" saved to: {output_path}")
```
